Loop control statement                                                
1.) Break statement-- this statement breaks the loop                  
2.) continue statement-- skip the current iteration                   
3.) pass statement-- only pass the code

```python
i=0
while i<5:
  i+=1
if i==3:
  break
else:
  print(i)
```

```python
i=0
while i<5:
  i+=1
  if i==3:
    continue
    print(i)
  else:
    print(i)
```

```python
i=0
while i<5:
  i+=1
  if i==3:
    pass
    print(i)
  else:
    print(i)
```

variable lenght argumnet:- we give any number of argument                      
1.) nonn keyword length argument:-  it behave like tuple and represented by star(*)                                                                           

2.) keywoard variable length argument:- it behaves like dictionry and represented by double star(**)

```python
#Non keyword variable length argument
def test(*args):
  print(args)
  print(len(args))
  print(type(args))
```

```python
test(1,2,3,4,5,6)
```

```python
def sum_number(*args):
  sum=0
  for num in args:
    sum+=num
  return sum
```

```python
print(sum_number(1,2,3,4,5,6))
```

```python
#keyword variable length argument
def test(**kwargs):
  print(kwargs)
  print(len(kwargs))
  print(type(kwargs))
```

```python
test(a=10,b=20,c=30)
```

```python
#How to plot
import matplotlib.pyplot as plt
import numpy as np
xpoint=np.array([1,2])
ypoint=np.array([1,2])
plt.plot(xpoint,ypoint)
plt.show()
```

```python
#Example
import matplotlib.pyplot as plt
import numpy as np
cgpa=np.array([6,7,8,9,10])
package=np.array([4,5,6,7,9])
plt.plot(cgpa,package,"*:g",ms=15,mec="red",mfc="orange",linewidth=5)
# * represent point
# : represent doted line for
# g represent the green color
# ms represent  the marker size
# mec marker edge color
# mfc marker face color
# line width to change the width of the line
plt.xlabel("CGPA")
plt.ylabel("Package")
plt.title("Shoolini university students data")
plt.grid()
# plt.grid() to show grid in the graph
plt.show()
```

```python
#Bar plot - it work on numerical as well as categorical data
import matplotlib.pyplot as plt
import numpy as np
fruit = np.array(["Apple","Banana","Orange","Watermelon"])
#to color
color= np.array(["Red","yellow","orange","green"])
quantity = np.array([100,120,125,134])
plt.barh(fruit,quantity,color=color)
plt.ylabel("Fruit")
plt.xlabel("Quantity")
plt.title("Fruit vs Quantity")
plt.show()

```

```python
#pie plot
import matplotlib.pyplot as plt
import numpy as np
percentage_of_fruit= np.array([100,120,125,134])
fruit = np.array(["Apple","Banana","Orange","Watermelon"])
#to explode a specfic thing
myexplode=np.array([0,0.2,0,0])
plt.pie(percentage_of_fruit,labels=fruit,explode=myexplode)
plt.legend(title="Fruit",loc="lower left")
plt.show()
```

```python
# to se data in dotted way:- use scatter plot
age_of_car=np.array([4,5,6,7,8,9,10,12,14])
speed_of_car=np.array([150,140,120,110,100,90,80,70,60])
plt.scatter(age_of_car,speed_of_car)

plt.xlabel("Age_of_car")
plt.ylabel("Speed_of_car")
plt.title("Age_of_car vs Speed_of_car")
plt.show()
```

```python
#histogram- it showing the frequency distribution
x = np.random.randint(40,100, (1,200))
# flatten is used to convert in 2D data into 1D
x=x.flatten()
print(x)
```

```python
# 18 november 2025 cosine similarity
import numpy as np
from numpy.linalg import norm
A = np.array([2,1,2,3,2,9])
B = np.array([3,4,2,4,5,5])
# compute cosine similarity
cosine = np.dot(A,B)/(norm(A)*norm(B))
print("Cosine Similarity:", cosine)
```
